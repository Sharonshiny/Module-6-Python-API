```python
#required libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import csv
import datetime as dt
from citipy import citipy as cy
import requests
import json
```

```python
#using random function to get 1000 the latitude and longitude values randomly
lat = list(np.random.uniform(-90, 91, 1000))
lat = [round(x,4)for x in lat]
lat[0:10]
```

```python
#get the random longitude values
lon = list(np.random.uniform(-180, 181, 1000))
lon = [round(x,4) for x in lon]
lon[0:10]
```

```python
#verify the number of items looped
len(lat), len(lon)
```

```python
#create a dataframe with latitude and longitude values
data1 = pd.DataFrame({"Latitude" : lat, "Longitude" : lon})
data1.head()
```

```python
#get the geographical details for each combination of latitude and longitude match the nearest city and the country it belongs to
data1["City"] = ""
data1["Country"] = ""

for idx, row in data1.iterrows():
    print(idx, row)
    city_list = cy.nearest_city(row["Latitude"], row["Longitude"])
    data1.set_value(idx, "City", city_list.city_name)
    data1.set_value(idx, "Country", city_list.country_code)
```

```python
#sample view of the mapped data1 dataframe
data1.head()
```

```python
#Get the unique values from the dataframe and drop if any duplicates present
unqData = data1.drop_duplicates()
```

```python
#number of datapoints in the form of rows and cols
unqData.shape
```

```python
#Adding datapoints to unqData dataframe from the weather api

unqData["Temperature"] = ""
unqData["Humidity"] = ""
unqData["Cloudiness"] = ""
unqData["Wind_Speed"] = ""

#key and endpoint assigned to the variables
endpoint = "http://api.openweathermap.org/data/2.5/weather?"
key = "15d3c7108387f4e9c4c2fd8727958398" 
c = 0

#calling the endpoint from api and printing the output from the restapi
for i,j in unqData.iterrows():
   c = c+1
   try:
       raw_data= requests.get(endpoint +"lat="+str(j[0]) +"&lon="+str(j[1])+"&appid="+key).json()
       unqData.set_value(i, "Temperature", raw_data["main"]["temp"])
       unqData.set_value(i, "Humidity", raw_data["main"]["humidity"])
       unqData.set_value(i, "Cloudiness", raw_data["clouds"]["all"])
       unqData.set_value(i, "Wind_Speed", raw_data["wind"]["speed"])
       print("The # {} City is {}:\n\t from the Country {}: \n\t\tand the current weather details are {}\n".format(c, j[2], j[3], raw_data))
   except:
       print("No data exists at this point of time for {}".format(j[2]))
```

```python
#sample view of the unqData dataframe
unqData.head()
```

```python
vizData = unqData[unqData["Temperature"]!= '']
vizData.shape
```

```python
#gives the data_type information and other properties of the dataframe
vizData.info()
```

```python
#convert the data types to integers or number for future ploting
vizData[["Temperature", "Humidity", "Cloudiness", "Wind_Speed"]] = vizData[["Temperature", "Humidity", "Cloudiness", "Wind_Speed"]].apply(pd.to_numeric)
```

```python
#look at the data
vizData.info()
vizData.head()
```

```python
#writing the dataframe to csv
vizData.to_csv("Weather_data.csv", index = False, encoding = 'utf-8')
```

```python
#get the date of plot at the time of execution
date = dt.datetime.today()
date =  date.strftime("%Y-%m-%d %H:%M")
date
```

```python
#visualization Temperature (F) vs. Latitude
plt.figure(figsize = (10,8))
plt.scatter(vizData["Latitude"], vizData["Temperature"], s = 50, c = 'red', edgecolors = 'blue')
plt.title("Temperature (F) vs. Latitude overview near Equatorial Region ({})".format(date), fontsize = 18)
plt.xlabel("Latitude", fontsize = 15)
plt.ylabel("Temperature(F)", fontsize = 15)
plt.axvline(x = 0, color = 'grey', alpha = 0.3)
plt.figtext(0.49, 0.2, "Equator", rotation = 90, fontsize = 10)
plt.savefig("Temperature_plot_along_Equator.png")
plt.show()
```

```python
#visualization Humidity (%) vs. Latitude
plt.figure(figsize = (10,8))
plt.scatter(vizData["Latitude"], vizData["Humidity"], s = 50, c = 'green', edgecolors = "red", alpha = 0.8)
plt.title("Humidity (%) vs. Latitude overview near Equatorial Region ({})".format(date), fontsize = 18)
plt.xlabel("Latitude", fontsize = 15)
plt.ylabel("Humidity(%)",fontsize = 15)
plt.axvline(x = 0, color = 'grey', alpha = 0.3)
plt.figtext(0.49, 0.20, "Equator", rotation = 90, fontsize = 10)
plt.savefig("Humidity_plot_along_Equator.png")
plt.show()
```

```python
#visualization Cloudiness (%) vs. Latitude
plt.figure(figsize = (10,8))
plt.scatter(vizData["Latitude"], vizData["Cloudiness"],  s = 70, edgecolors = 'pink')
plt.title("Cloudiness (%) vs. Latitude overview near Equatorial Region({})".format(date), fontsize = 18)
plt.ylabel("Cloudiness(%)", fontsize = 15)
plt.xlabel("Latitude", fontsize = 15)
plt.axvline(x = 0, color = 'grey', alpha = 0.3)
plt.figtext(0.49, 0.85, "Equator", rotation = 90, fontsize = 10)
plt.savefig("Cloudiness_plot_along_Equator.png")
plt.show()
```

```python
#visualization Wind Speed (mph) vs. Latitude
plt.figure(figsize = (10,8))
plt.scatter(vizData["Latitude"], vizData["Wind_Speed"], s = 100, c = 'green', edgecolors = 'blue')
plt.title("Wind_Speed (mph) vs. Latitude overview near Equatorial Region ({})".format(date), fontsize = 18)
plt.ylabel("Wind_Speed(mph)", fontsize = 15)
plt.xlabel("Latitude",fontsize = 15)
plt.axvline(x = 0, color = 'grey', alpha = 0.3)
plt.figtext(0.49, 0.85, "Equator", rotation = 90, fontsize = 10)
plt.savefig("WindSpeed_plot_along_Equator.png")
plt.show()
```
